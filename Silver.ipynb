# Merge using pyspark

```sql
%python
if spark.catalog.tableExists("datamodelling_lak.silver.silver_table"):
    pass
else:
    spark.sql(f"""
              create table if not exists datamodelling_lak.silver.silver_table 
              as 
              select * from silver_source""")
```

# Merge using sql

```sql
merge into datamodelling_lak.silver.silver_table
using silver_source
on datamodelling_lak.silver.silver_table.order_id = silver_source.order_id
when matched then update set *
when not matched then insert *
```
