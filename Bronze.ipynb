```sql
%python
last_load_date = '1000-01-01'
```

```sql
%python
if spark.catalog.tableExists("datamodelling_lak.bronze.bronze_table"):
    last_load_date = spark.sql(f""" select max(order_date) from datamodelling_lak.bronze.bronze_table""").collect()[0][0]
else:
    last_load_date = '1000-01-01'
```

```sql
%python
last_load_date
```

```sql
%python
spark.sql(f"""select * from datamodelling_lak.default.source_data
where order_date > '{last_load_date}'""").createOrReplaceTempView("Bronze_source")
```

```sql
create or replace table datamodelling_lak.bronze.bronze_table as
select * from bronze_source
```
